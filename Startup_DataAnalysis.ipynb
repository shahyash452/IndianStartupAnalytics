```python
df['Industry Vertical']=df['Industry Vertical'].fillna(value='Other')
df['SubVertical']=df['SubVertical'].fillna(value='Other')
df['City  Location']=df['City  Location'].fillna(value='Other')
```

```python
df.isnull().sum()
```

```python
newdf=df.copy()
del newdf['Remarks']
del newdf['Sr No']
```

```python
newdf
```

```python
import re
def convert_AmountInUSD(x):
    if re.search(',',x):
        return (x.replace(',',''))
    return x
```

```python
newdf['Amount in USD']=newdf[newdf['Amount in USD'].notnull()]['Amount in USD'].apply(convert_AmountInUSD)
newdf
```

```python
newdf['Amount in USD']=newdf['Amount in USD'].fillna(0)
newdf
```

```python
newdf['Amount in USD']=newdf['Amount in USD'].replace('Undisclosed',np.nan)
newdf['Amount in USD']=newdf['Amount in USD'].replace('Unknown',np.nan)
newdf['Amount in USD']=newdf['Amount in USD'].replace('undisclosed',np.nan)
newdf['Amount in USD']=newdf['Amount in USD'].replace('unknown',np.nan)
```

```python
newdf
```

```python
newdf
```

```python
newdf['Amount in USD']=newdf['Amount in USD'].astype('float')
```

```python
newdf['Amount in USD']=round(newdf['Amount in USD'].replace(0,(np.mean(newdf['Amount in USD']))))
```

```python
newdf
```

```python
newdf.isnull().sum()
```

```python
topTenFunding=show(newdf.groupby('Startup Name')['Amount in USD'].sum().sort_values(ascending=False))
topTenFunding.head(5)
```

```python
d=newdf['Industry Vertical'].value_counts().head(5)
f=newdf.groupby('InvestmentnType').sum()['Amount in USD']
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(20,10))
labels=[d.index,f.index]
size=[d.values,f.values]
colors = [['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','pink'],['green','pink','red','yellow']]
plt.axis('equal')
explode = ((0.1, 0, 0, 0,0),(-0.5,0.5,0.1,0.1))
ax[0].pie(size[0],explode=explode[0], labels=labels[0], colors=colors[0],
        autopct='%1.1f%%', shadow=True, startangle=140)

ax[1].pie(size[1],explode=explode[1], labels=labels[1], colors=colors[1],autopct='%1.5f%%', shadow=True, startangle=140)
plt.show()
```

```python
plt.figure(figsize=(17,12))
mean_amount = newdf.groupby('City  Location').mean()["Amount in USD"].astype('int').sort_values(ascending=False).iloc[1:].head(10)
squarify.plot(sizes=mean_amount.values,label=mean_amount.index, value=mean_amount.values,color=['crimson','seagreen','olive','hotpink','deepskyblue','grey','purple','lime','yellow','orange'])
plt.title('Distribution of Startups across Top cities')
```
